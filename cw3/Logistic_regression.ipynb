```python
import pandas as pd
crus = pd.read_csv("VLagun_Crus_Basins.csv")
rotif = pd.read_csv("VLagun_Rotif_Basins.csv")
crus.head()
```

```python
rotif.head()
```

```python
crus.dropna()
```

```python
rotif.dropna()
```

```python
X_crus = pd.get_dummies(crus)
X_rotif = pd.get_dummies(rotif)
X_crus.drop(["Basin"], inplace=True, axis=1)
X_rotif.drop(["Basin"], inplace=True, axis=1)
Y_crus = crus['Basin']
Y_rotif = rotif['Basin']
```

```python
from sklearn.model_selection import train_test_split
X_crus = X_crus.iloc[:,0:12]
X_rotif = X_rotif.iloc[:,0:19]
Y_crus = Y_crus.iloc[:,]
Y_rotif = Y_rotif.iloc[:,]
test_size = 0.25
X_c_train, X_c_test, Y_c_train, Y_c_test = train_test_split(X_crus, Y_crus, test_size= test_size, random_state= 16)
print("Shape of X_train:", X_c_train.shape)
print("Shape of X_test:", X_c_test.shape)
```

```python
X_r_train, X_r_test, Y_r_train, Y_r_test = train_test_split(X_rotif, Y_rotif, test_size= test_size, random_state= 16)
print("Shape of X_train:", X_r_train.shape)
print("Shape of X_test:", X_r_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
logreg_c = LogisticRegression(random_state= 16)
logreg_r = LogisticRegression(random_state= 16)
logreg_c.fit(X_c_train, Y_c_train)
Y_c_pred = logreg_c.predict(X_c_test)
logreg_r.fit(X_r_train, Y_r_train)
Y_r_pred = logreg_r.predict(X_r_test)
```

```python
from sklearn import metrics

cnf_c_matrix = metrics.confusion_matrix(Y_c_test, Y_c_pred)
cnf_c_matrix
```

```python
cnf_r_matrix = metrics.confusion_matrix(Y_r_test, Y_r_pred)
cnf_r_matrix
```

```python
  import numpy as np
  import matplotlib.pyplot as plt
  import seaborn as sns

  class_names=[0,1]
  fix, ax = plt.subplots()
  tick_marks = np.arange(len(class_names))
  plt.xticks(tick_marks, class_names)
  plt.yticks(tick_marks, class_names)

  sns.heatmap(pd.DataFrame(cnf_c_matrix), annot=True, cmap="YlGnBu", fmt='g')
  ax.xaxis.set_label_position("top")
  plt.tight_layout()
  plt.title("Confucion Matrix of Crus", y=1.1)
  plt.ylabel("Actual label")
  plt.xlabel("Predicted label")
  plt.show()
```

```python
from sklearn.metrics import classification_report

target_names = ["Influence of sea waters", "Influence of land waters"]
print("================================== CRUS ==================================")
print(classification_report(Y_c_test, Y_c_pred, target_names=target_names))
```

```python
print("================================= ROTIF =================================")
print(classification_report(Y_r_test, Y_r_pred, target_names=target_names))
```

```python
Y_c_pred_proba = logreg_c.predict_proba(X_c_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(Y_c_test, Y_c_pred_proba)
auc = metrics.roc_auc_score(Y_c_test, Y_c_pred_proba)
plt.plot(fpr, tpr, label='data 1, auc = '+str(auc))
plt.legend(loc=4)
plt.show()
```
