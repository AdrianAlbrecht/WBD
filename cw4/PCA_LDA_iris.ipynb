```python
import matplotlib.pyplot as plt

from sklearn import datasets
from sklearn.decomposition import PCA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
iris= datasets.load_iris()
```

```python
X = iris.data
Y = iris.target
target_names = iris.target_names
```

```python
pca = PCA(n_components=2)
X_r =  pca.fit(X).transform(X)
```

```python
lda = LinearDiscriminantAnalysis(n_components=2)
X_r2 = lda.fit(X, Y).transform(X)
```

```python
print(f"explained variance ratio (first two components in PCA): {str(pca.explained_variance_ratio_)}")
```

```python
print(f"explained variance ratio (first two components in LDA): {str(lda.explained_variance_ratio_)}")
```

```python
plt.figure()

colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0,1,2], target_names):
  plt.scatter(X_r[Y==i,0],X_r[Y==i,1], color=color, alpha = .8, lw = lw, label=target_name)

plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("PCA of IRIS dataset")
```

```python
plt.figure()

colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0,1,2], target_names):
  plt.scatter(X_r2[Y==i,0],X_r2[Y==i,1], color=color, alpha = .8, lw = lw, label=target_name)

plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.title("LDA of IRIS dataset")
```
